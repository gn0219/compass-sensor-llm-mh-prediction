````python
# Additional imports for analysis
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
import os
from tqdm import tqdm
import re
````

````python
# Configuration
N_BOOTSTRAP = 1000  # Number of bootstrap iterations (set to 0 to skip bootstrapping)
OUTPUT_DIR = '/home/iclab/compass/compass-sensor-llm-mh-prediction/supplementary_analysis/llm_final_results'
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Note:
# - To generate summary CSV WITHOUT bootstrapping (only basic metrics), set N_BOOTSTRAP = 0
# - To generate summary CSV WITH confidence intervals, set N_BOOTSTRAP = 1000 (or any positive number)
````

````python
def parse_filename(filepath):
    """
    Parse the filename to extract metadata.
    Example: ces_compass_4shot_personalrecent_selffeedback_seed42_gpt_5_self_feedback_42_20251031_153634.json
    """
    filename = os.path.basename(filepath)
    parts = filename.replace('.json', '').split('_')
    
    # Extract dataset (ces, globem, mentaliot)
    dataset = parts[0]
    
    # Extract n_shot (zeroshot or 4shot)
    if 'zeroshot' in filename:
        n_shot = '0-shot'
    elif '4shot' in filename:
        n_shot = '4-shot'
    else:
        n_shot = 'unknown'
    
    # Extract ICL strategy
    if 'personalrecent' in filename:
        icl_strategy = 'Personal Recent'
    elif 'crossrandom' in filename:
        icl_strategy = 'Cross Random'
    elif 'crossretrieval' in filename:
        icl_strategy = 'Cross Retrieval'
    elif 'hybridblend' in filename:
        icl_strategy = 'Hybrid'
    elif 'none' in filename:
        icl_strategy = 'None'
    else:
        icl_strategy = 'unknown'
    
    # Extract reasoning
    if 'selffeedback' in filename:
        reasoning = 'Self-Feedback'
    elif 'cot' in filename:
        reasoning = 'CoT'
    elif 'direct' in filename:
        reasoning = 'Direct'
    else:
        reasoning = 'unknown'
    
    # Extract model name
    if 'gpt_5' in filename:
        model = 'GPT-5'
    elif 'gpt_oss_20b' in filename:
        model = 'gpt-oss-20b'
    elif 'claude_4_5_sonnet' in filename:
        model = 'Claude Sonnet 4.5'
    elif 'gemini_2_5_pro' in filename:
        model = 'Gemini 2.5 Pro'
    elif 'llama_3_1_8b' in filename:
        model = 'Llama-3.1 8b'
    elif 'mistral_7b' in filename:
        model = 'Mistral 7B'
    else:
        model = 'unknown'
    
    return {
        'dataset': dataset,
        'N_shot': n_shot,
        'ICL_strategy': icl_strategy,
        'reasoning': reasoning,
        'model': model
    }
````

````python
def convert_prediction_to_binary(pred_value):
    """
    Convert 'Low Risk' to 0, 'High Risk' to 1, others to None (error case)
    """
    if pred_value == 'Low Risk':
        return 0
    elif pred_value == 'High Risk':
        return 1
    else:
        return None  # Error case
````

````python
def calculate_classification_metrics(predictions, ground_truth_df, dataset):
    """
    Calculate classification metrics from predictions.
    Returns a dictionary with metrics for each condition.
    """
    # Create a DataFrame from predictions
    pred_data = []
    for pred in predictions:
        user_id = pred['user_id']
        ema_date = pred['ema_date']
        
        # Get prediction values from 'prediction' field
        pred_dict = pred.get('prediction', {}).get('Prediction', {})
        
        anxiety_pred = convert_prediction_to_binary(pred_dict.get('Anxiety'))
        depression_pred = convert_prediction_to_binary(pred_dict.get('Depression'))
        
        # Stress is not available for globem
        if dataset == 'globem':
            stress_pred = None
        else:
            stress_pred = convert_prediction_to_binary(pred_dict.get('Stress'))
        
        pred_data.append({
            'user_id': user_id,
            'ema_date': ema_date,
            'pred_anxiety': anxiety_pred,
            'pred_depression': depression_pred,
            'pred_stress': stress_pred
        })
    
    pred_df = pd.DataFrame(pred_data)
    
    # Merge with ground truth
    merged_df = ground_truth_df.merge(pred_df, on=['user_id', 'ema_date'], how='inner')
    
    # Remove error cases (where prediction is None)
    merged_df_anxiety = merged_df[merged_df['pred_anxiety'].notna()].copy()
    merged_df_depression = merged_df[merged_df['pred_depression'].notna()].copy()
    
    if dataset != 'globem':
        merged_df_stress = merged_df[merged_df['pred_stress'].notna()].copy()
    # Calculate metrics
    metrics = {}

    metric_names = ['accuracy', 'f1_macro', 'precision', 'recall']
    conditions = [
        ('depression', merged_df_depression, 'Depression', 'pred_depression'), 
        ('anxiety', merged_df_anxiety, 'Anxiety', 'pred_anxiety'),
    ]
    if dataset != 'globem':
        conditions.append(('stress', merged_df_stress, 'Stress', 'pred_stress'))

    metric_funcs = {
        'accuracy': accuracy_score,
        'f1_macro': lambda y_true, y_pred: f1_score(y_true, y_pred, average='macro', zero_division=0),
        'precision': lambda y_true, y_pred: precision_score(y_true, y_pred, average='macro', zero_division=0),
        'recall': lambda y_true, y_pred: recall_score(y_true, y_pred, average='macro', zero_division=0)
    }

    for cond_name, df, col_true, col_pred in conditions:
        if len(df) > 0:
            y_true = df[col_true]
            y_pred = df[col_pred]
            for metric_name in metric_names:
                metrics[f"{cond_name}_{metric_name}"] = metric_funcs[metric_name](y_true, y_pred)
        else:
            for metric_name in metric_names:
                metrics[f"{cond_name}_{metric_name}"] = None
    return metrics, merged_df
````

````python
def bootstrap_metrics(merged_df, dataset, n_bootstrap=1000, seed=42):
    """
    Perform bootstrapping to calculate confidence intervals for metrics.
    If n_bootstrap is 0, skip bootstrapping and return empty dict.
    """
    if n_bootstrap == 0:
        return {}
    
    np.random.seed(seed)
    
    conditions = ['depression', 'anxiety']
    if dataset != 'globem':
        conditions.append('stress')
    
    metrics = ['accuracy', 'f1_macro', 'precision', 'recall']
    
    bootstrap_results = {f'{cond}_{metric}': [] for cond in conditions for metric in metrics}
    
    for _ in range(n_bootstrap):
        # Bootstrap sample
        sample_df = merged_df.sample(n=len(merged_df), replace=True)
        
        # Calculate metrics for each condition
        for condition in conditions:
            # Filter out error cases for this condition
            condition_df = sample_df[sample_df[f'pred_{condition}'].notna()].copy()
            
            if len(condition_df) > 0:
                y_true = condition_df[condition.capitalize()]
                y_pred = condition_df[f'pred_{condition}']
                
                bootstrap_results[f'{condition}_accuracy'].append(
                    accuracy_score(y_true, y_pred)
                )
                bootstrap_results[f'{condition}_f1_macro'].append(
                    f1_score(y_true, y_pred, average='macro', zero_division=0)
                )
                bootstrap_results[f'{condition}_precision'].append(
                    precision_score(y_true, y_pred, average='macro', zero_division=0)
                )
                bootstrap_results[f'{condition}_recall'].append(
                    recall_score(y_true, y_pred, average='macro', zero_division=0)
                )
    
    # Calculate 95% confidence intervals
    ci_results = {}
    for key, values in bootstrap_results.items():
        if len(values) > 0:
            ci_lower = np.percentile(values, 2.5)
            ci_upper = np.percentile(values, 97.5)
            ci_results[f'{key}_ci'] = f'[{ci_lower:.3f}, {ci_upper:.3f}]'
        else:
            ci_results[f'{key}_ci'] = None
    
    return ci_results
````

````python
def process_dataset(dataset_name, result_files, ground_truth_dfs, n_bootstrap=1000):
    """
    Process all result files for a dataset and create summary CSV.
    """
    print(f"\nProcessing {dataset_name.upper()} dataset...")

    all_results = []

    for file_path in tqdm(sorted(result_files), desc=f"Processing {dataset_name}"):
        try:
            # Parse filename to get metadata
            metadata = parse_filename(file_path)

            if metadata['dataset'] != dataset_name:
                continue

            # Load JSON file
            with open(file_path, 'r') as f:
                data = json.load(f)

            # Get ground truth for this dataset
            gt_df = ground_truth_dfs[dataset_name]

            # Calculate classification metrics from predictions
            predictions = data.get('predictions', [])
            class_metrics, merged_df = calculate_classification_metrics(
                predictions, gt_df, dataset_name
            )

            # Get cost efficiency metrics
            if 'self_feedback' in file_path:
                cost_eff = data.get('cost_efficiency', {}).get('overall', {})
            else:
                cost_eff = data.get('cost_efficiency', {})

            # Extract tokens & efficiency metrics
            tokens_metrics = cost_eff.get('tokens', {})
            latency_metrics = cost_eff.get('latency', {})
            cost_metrics = cost_eff.get('cost', {})

            # Perform bootstrapping
            ci_results = bootstrap_metrics(merged_df, dataset_name, n_bootstrap=n_bootstrap)

            # Combine all metrics
            result = {
                # Metadata
                'N_shot': metadata['N_shot'],
                'ICL_strategy': metadata['ICL_strategy'],
                'reasoning': metadata['reasoning'],
                'model': metadata['model'],

                # Classification Performance
                'depression_accuracy': class_metrics.get('depression_accuracy'),
                'depression_f1_macro': class_metrics.get('depression_f1_macro'),
                'depression_precision': class_metrics.get('depression_precision'),
                'depression_recall': class_metrics.get('depression_recall'),

                'anxiety_accuracy': class_metrics.get('anxiety_accuracy'),
                'anxiety_f1_macro': class_metrics.get('anxiety_f1_macro'),
                'anxiety_precision': class_metrics.get('anxiety_precision'),
                'anxiety_recall': class_metrics.get('anxiety_recall'),
            }

            # Add stress metrics if not globem
            if dataset_name != 'globem':
                result.update({
                    'stress_accuracy': class_metrics.get('stress_accuracy'),
                    'stress_f1_macro': class_metrics.get('stress_f1_macro'),
                    'stress_precision': class_metrics.get('stress_precision'),
                    'stress_recall': class_metrics.get('stress_recall'),
                })

            # Add Tokens & Efficiency metrics
            result.update({
                'prompt_avg': tokens_metrics.get('prompt_avg'),
                'prompt_std': tokens_metrics.get('prompt_std'),
                'completion_avg': tokens_metrics.get('completion_avg'),
                'completion_std': tokens_metrics.get('completion_std'),
                'avg_per_request': tokens_metrics.get('avg_per_request'),
                'std_per_request': tokens_metrics.get('std_per_request'),
                'avg_seconds': latency_metrics.get('avg_seconds'),
                'std_seconds': latency_metrics.get('std_seconds'),
                'per_sample_usd': cost_metrics.get('per_sample_usd'),
                'per_sample_std': cost_metrics.get('per_sample_std'),
            })

            # Add bootstrap confidence intervals
            result.update(ci_results)

            all_results.append(result)

        except Exception as e:
            print(f"Error processing {file_path}: {str(e)}")
            continue

    # Create DataFrame
    df_results = pd.DataFrame(all_results)

    # -- Custom COLUMN and ROW order --
    # 1. Sort columns: ICL_strategy, reasoning, model (with custom model order), then the rest
    priority_cols = ['N_shot', 'ICL_strategy', 'reasoning', 'model']

    n_shot_order = ['0-shot', '4-shot']
    icl_strategy_order = ['None', 'Cross Random', 'Cross Retrieval', 'Personal Recent', 'Hybrid']
    reasoning_order = ['Direct', 'CoT', 'Self-Feedback']
    model_order = [
        "GPT-5", 
        "Claude Sonnet 4.5", 
        "Gemini 2.5 Pro", 
        "Llama-3.1 8b", 
        "Mistral 7B", 
        "gpt-oss-20b"
    ]

    # ì •ë ¬ì„ ìœ„í•œ rank í•¨ìˆ˜ë“¤ ì •ì˜
    def get_rank_value(val, order):
        val_str = (val or "").lower()
        for idx, item in enumerate(order):
            # ëª¨ë¸ì€ ë¶€ë¶„ ë§¤ì¹˜ (in), ë‹¤ë¥¸ ì¹´í…Œê³ ë¦¬ëŠ” ì™„ì „ ì¼ì¹˜
            if order is model_order:
                if item.lower() in val_str:
                    return idx
            else:
                if val_str == item.lower():
                    return idx
        return len(order)

    # ì •ë ¬ ê¸°ì¤€ ì»¬ëŸ¼ ì¶”ê°€
    df_results['N_shot_rank'] = df_results['N_shot'].apply(lambda v: get_rank_value(v, n_shot_order))
    df_results['ICL_strategy_rank'] = df_results['ICL_strategy'].apply(lambda v: get_rank_value(v, icl_strategy_order))
    df_results['reasoning_rank'] = df_results['reasoning'].apply(lambda v: get_rank_value(v, reasoning_order))
    df_results['model_rank'] = df_results['model'].apply(lambda v: get_rank_value(v, model_order))

    # Make sure missing columns are filled (in case)
    other_cols = [c for c in df_results.columns if c not in priority_cols and not c.endswith("_rank")]
    col_order = priority_cols + other_cols

    # 2. Sort rows by rank columns
    df_results = df_results.sort_values(
        by=['N_shot_rank', 'ICL_strategy_rank', 'reasoning_rank', 'model_rank']
    ).reset_index(drop=True)

    # rank ì»¬ëŸ¼ ì‚­ì œ
    df_results = df_results.drop(columns=[c for c in df_results.columns if c.endswith('_rank')])

    # ì»¬ëŸ¼ ìˆœì„œ ì ìš©
    df_results = df_results[col_order]

    # Save to CSV
    output_path = os.path.join(OUTPUT_DIR, f'{dataset_name}_summary.csv')
    df_results.to_csv(output_path, index=False)
    print(f"Saved {dataset_name} results to {output_path}")

    return df_results
````

````python
# Load ground truth labels for all datasets
def load_ground_truth(dataset_name):
    """
    Load ground truth labels for a dataset.
    """
    # Find the first matching directory for this dataset
    saved_prompts_dir = '/home/iclab/compass/compass-sensor-llm-mh-prediction/saved_prompts_old/saved_prompts'
    
    # Get all subdirectories
    subdirs = [d for d in os.listdir(saved_prompts_dir) 
               if os.path.isdir(os.path.join(saved_prompts_dir, d)) and d.startswith(dataset_name)]
    
    if not subdirs:
        raise ValueError(f"No directory found for dataset {dataset_name}")
    
    # Use the first matching directory
    label_path = os.path.join(saved_prompts_dir, subdirs[0], 'labels.json')
    
    with open(label_path, 'r') as f:
        labels = json.load(f)
    
    if dataset_name == 'globem':
        # Globem doesn't have stress
        df = pd.DataFrame([
            {
                'user_id': l['user_id'],
                'ema_date': l['ema_date'],
                'Anxiety': l['true_anxiety'],
                'Depression': l['true_depression'],
            }
            for l in labels
        ])
    else:
        df = pd.DataFrame([
            {
                'user_id': l['user_id'],
                'ema_date': l['ema_date'],
                'Anxiety': l['true_anxiety'],
                'Depression': l['true_depression'],
                'Stress': l['true_stress'],
            }
            for l in labels
        ])
    
    return df

print("Loading ground truth labels...")
ground_truth_dfs = {
    'ces': load_ground_truth('ces'),
    'globem': load_ground_truth('globem'),
    'mentaliot': load_ground_truth('mentaliot')
}

print(f"CES: {len(ground_truth_dfs['ces'])} samples")
print(f"GLOBEM: {len(ground_truth_dfs['globem'])} samples")
print(f"MENTALIOT: {len(ground_truth_dfs['mentaliot'])} samples")
````
